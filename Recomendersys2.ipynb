```python
import ast
def func1(gen):
    li=[]
    for i in ast.literal_eval(gen):
        li.append(i['name'])
    return li    
    
```

```python
func1('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movie['genres']=movie['genres'].apply(func1)
```

```python
movie.head(2)
```

```python
movie["keywords"][0]
```

```python
movie['keywords'] = movie['keywords'].apply(func1)
movie.head()
```

```python
movie.cast[0]

```

```python
#we have to take top 3 casts
def func2(cas):
    li=[]
    count=0
    for i in ast.literal_eval(cas):
        if count<3:
            li.append(i['name'])
            count+=1
        else:
            break
    return li        
        
```

```python
movie["cast"]=movie['cast'].apply(func2)
```

```python
movie.sample()
```

```python
movie["crew"][0]
```

```python
# from crew fetch director name only
def get_director(ele):
    li=[]
    for i in ast.literal_eval(ele):
        if i['job'] == 'Director':
            li.append(i['name'])
    return li
        
```

```python
movie['crew']=movie["crew"].apply(get_director)
```

```python
movie.head()
```

```python
def remove_space(text):
    li=[]
    for i in text:
        li.append(i.replace(" ",""))
    return li    
```

```python
movie['keywords']=movie.keywords.apply(remove_space)
```

```python
movie.head(2)
```

```python
movie['cast']=movie.cast.apply(remove_space)
movie['crew']=movie.crew.apply(remove_space)
movie['genres']=movie.crew.apply(remove_space)
```

```python
movie.sample(4)
```

```python
movie.isnull().sum()
```

```python
movie.dropna(inplace=True)
```

```python
movie.isnull().sum()
```

```python
movie.info()
```

```python
#concat the genres,overview,keyword,cast,crew in one column called tages
movie['overview']=movie["overview"].apply(lambda x:x.split())
```

```python
movie["overview"]
```

```python
movie["tages"]=movie['overview']+movie['genres']+movie["keywords"]+movie["cast"]+movie["crew"]
```

```python
movie["tages"]
```

```python
movie.drop(columns=['overview','genres','keywords','cast','crew'],inplace=True)
```

```python
movie.sample(2)
```

```python
movie["tages"]=movie["tages"].apply(lambda x:" ".join(x))
```

```python
movie.sample(2)
```

```python
#use count vectorizer of bag of words for converting into string to number
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vector=cv.fit_transform(movie["tages"]).toarray()
```

```python
vector
```

```python
cv.vocabulary_
```

```python
#we have to use cosine_similirity for getting minimam distance between two vectorr so that we can
#group two movie according to distance or angle
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar=cosine_similarity(vector)
```

```python
similar[0]
```

```python
def recomanded_movie(mov):
    index=movie[movie["title"]==mov].index[0]
    distance=sorted(list(enumerate(similar[index])),reverse=True,key=(lambda x:x[1]))
    for i in distance[1:6]:
        print(movie.iloc[i[0]].title)
```

```python
recomanded_movie('Gandhi')
```
